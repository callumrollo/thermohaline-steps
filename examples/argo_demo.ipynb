# Argo demo

We can use the classifer on an Argo profile fetched with argopy

```python
from argopy import DataFetcher as ArgoDataFetcher
import matplotlib.pyplot as plt
import gsw
import xarray as xr
from thermohaline import detect_staircases, staircase_functions
```

```python
# float profiles from a specific region
ds = ArgoDataFetcher().region([-58, -56, 12, 14, 0, 1000, '2025-05', '2026-05']).to_xarray()
ds = ds.argo.point2profile()
ds_argo =  ds.sel({'N_PROF':3})
```

```python
# The example profile From vanderBoog paper. Argo float 6901769 at 8.9◦ E, 37.9◦ N on 31 October 2017. Profile 170
ds_argo = ArgoDataFetcher().profile(6901769, 170).to_xarray()
```

### Interpolate

We interpolate the argo data onto an even pressure strep for use with the classifier

```python
ds_interp = staircase_functions.interp_ds(ds_argo, pressure_step=1, plot=True)
```

### Derived variables

We use gsw in the utility function `calc_derived_variables` to calculate conservative temperature and absolute salinity

```python
ds = staircase_functions.calc_derived_variables(ds_interp)
```

### Detect staircases

```python
df, mixes, grads = detect_staircases.classify_staircase(ds.PRES.values, ds.CTEMP.values, ds.ASAL.values, show_steps=True)
```
